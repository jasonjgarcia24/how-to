# Set Up Your Python Virtual Environment:

To install the conda virtual environment, follow these steps:

1. Open Terminal (macOS) or Git Bash (Windows).

2. <b>macOS users:</b> If you're running an OS version earlier than 10.15 OR running version 10.15 with a `bash` terminal environment, type `conda init bash`. If you're running versions 10.15 (Catalina) or later AND running the `zsh` environment, type `conda init zsh`.

   <b>Windows users:</b> Type `conda init bash` in Git Bash.

3. Close and reopen Terminal (macOS) or Git Bash (Windows) for the change to take effect.

4. Type `conda --version` at the command prompt to check the `conda` version that is installed.

5. Update conda by running the update command `conda update conda`. If a newer version is available, accept the update by typling "y".

6. To create the `conda` virtual environment, type the following: `conda create -n dev python=3.7 anaconda`. When prompted, choos `y` to proceed. Something like the following image should appear in the terminal window:



<b>Note:</b>
You'll use the instructions `conda activate dev` and `conda deactivate` frequently when initiating and disabling the conda dev environments on your machine

7. In the Git Bash terminal, type `conda init bash` to ready the terminal for conda environments. Close and reopen Git Bash, and then enter `conda activate dev` to activate the development environment.
