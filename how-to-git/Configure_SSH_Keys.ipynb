#### Adding Default Keys:

In the simplest form, ssh-add can be run without arguments. Used this way, it adds the default keys ~/.ssh/id_rsa, ~/.ssh/id_dsa, ~/.ssh/id_ecdsa, ~/.ssh/id_ed25519, and ~/.ssh/identity, if they exits.

#### Adding an Arbitrary Key:

To add an arbitratry private key, give the path of the key file as an argument to `ssh-add`. For example:<br>
`ssh-add ~/.ssh/apple-key`<br><br>
This would add the files ~/.ssh/apple-key and ~/.ssh/apple-key.pub.

# 4) Add SSH Key to GitLab:

To add the key to <a href="https://ucb.bootcampcontent.com/UCB-Coding-Bootcamp">GitLab</a>, copy the key to your clipboard by entering `clip < ~/.ssh/id_rsa.pub`. This copies the contents of the public key to your clipboard.<br><br>
In GitLab, go to User Settings>>SSH Keys and paste the public key contents into the "Key" field. Then add a descriptive title in the "Title" field and press the "Add key" pushbutton.

#### Error: Permission denied (publickey)
A "Permission denied" error means that th server rejected your connection. There could be several reasons why, and the most common examples are explained below.

##### Check that you are connecting to the correct server:
To make sure you are connecting to the right domain, you can enter the following command:<br>
<code>ssh T git@github.com</code>

The connection should be made on port 22, unless you're overriding the settings to use SSH over HTTPS.

##### Always user the "git" user:
All connections, including those for remote URLs, must be made as the "git" user. If you try to connect with your GitHub username, it will fail:<br>
`ssh -T GITHUB-USERNAME@github.com`
