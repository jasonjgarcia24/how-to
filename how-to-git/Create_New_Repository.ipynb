# Create New Repository (Repo)

1. On <a href="https://github.com/">github.com</a> and in the `Repositories` panel, select `New`.<br>


2. Settings:
 - A `.gitignore` file allows you to populate filenames and file extensions that you do not wish to be included in your commits to the repository.<br><br>A common use for this file is to prevent `config` files that have api keys or passwords stored in them from making their way into your repository.<br><br>`.gitignore` files come prepopulated with certain files and extension names specific to your programming language, but can be customized to fit your needs.
