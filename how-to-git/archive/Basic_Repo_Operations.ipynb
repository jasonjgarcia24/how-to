# Clone GitHub/GitLab Repository
1. Identify the repository to clone:
 - GitHub: `https://github.com/jasonjgarcia24/codingdemo`
 - GitLab: `git@ucb.bootcampcontent.com:UCB-Coding-Bootcamp/ucb-virt-fin-pt-06-2021-u-b.git`
2. Copy the repo link.
3. In the terminal, cd into the directory where you want your local repo to live: `cd ~/Fintech`
4. Clone the repo: `git clone https://github.com/jasonjgarcia24/codingdemo`

# Push GitHub Repository
1. Use `cd ~/Fintech/ucb-virt-fin-pt-06-2021-u-b` to move to local repo.
2. Make change.
3. Use `git add .` in Bash window to add all the files to the given folder.
4. Use `git status` in Bash window to view all the files which are going to be staged to the first commit.
5. Use `git commit -m 'your message` to commit the files staged in your local repo.
 - The `-m` flag is to enable a message to set the commits with the content where the full description is included, and a message is written in an imperative sentence up to 50 characters long and defining "what was changed", and "why was the change made".
6. Ensure "origin" is pointed to the correct location. For example: `git remote set-url origin git@jasonjgarcia24`
6. Use `git push -u origin main` to push the update.<br><br>
